```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HarshithRR29/Assignment-1/main/datasets/StudentsPerformance.csv')
df.head()
```

## Visualization 1: Distribution of Math Scores

```python
plt.figure(figsize=(10, 6))
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```

##  Visualization 2: Reading Scores by Gender

```python
plt.figure(figsize=(10, 6))
df.groupby('gender')['reading score'].mean().plot(kind='bar', color=['lightcoral', 'lightskyblue'])
plt.title('Average Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Reading Score')
plt.grid(True)
plt.show()
```

## Visualization 3: Writing Scores by Parental Level of Education

```python
plt.figure(figsize=(12, 8))
df.groupby('parental level of education')['writing score'].mean().plot(kind='bar', color='lightgreen')
plt.title('Average Writing Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Writing Score')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

## Visualization 4: Math Scores by Gender

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='math score', by='gender', grid=False, patch_artist=True, boxprops=dict(facecolor='lightyellow'))
plt.title('Math Scores by Gender')
plt.suptitle('')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Visualization 5: Correlation Heatmap of Scores

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Scores')
plt.show()
```
